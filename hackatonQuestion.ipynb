```python
# Importer les bibliothèques nécessaires pour l'analyse (panda, json et xml pour lire les fichiers)
import pandas as pd
import json
import xml.etree.ElementTree as ET

# Charger et préparer les données JSON
with open('/content/hackaton/olympic_athletes.json', 'r') as file:
    athletes_data = json.load(file)

df_athletes = pd.DataFrame(athletes_data)

# Charger et préparer les données HTML
df_results = pd.read_html('/content/hackaton/olympic_results.html')[0]

# Charger et préparer les données XML
tree = ET.parse('/content/hackaton/olympic_hosts.xml')
root = tree.getroot()

data = []
for elem in root:
    record = {}
    for subelem in elem:
        record[subelem.tag] = subelem.text
    data.append(record)

df_hosts = pd.DataFrame(data)

# Charger et préparer les données Excel
df_medals = pd.read_excel('/content/hackaton/olympic_medals.xlsx')
```

```python
# Nettoyage des noms de colonnes
df_athletes.columns = df_athletes.columns.str.strip().str.lower().str.replace(' ', '_')
df_results.columns = df_results.columns.str.strip().str.lower().str.replace(' ', '_')
df_hosts.columns = df_hosts.columns.str.strip().str.lower().str.replace(' ', '_')
df_medals.columns = df_medals.columns.str.strip().str.lower().str.replace(' ', '_')

print(df_athletes.columns)
print(df_results.columns)
print(df_hosts.columns)
print(df_medals.columns)
```

```python
# Question 1

# Filtre sur les jeux organisés par la France
df_hosts_france = df_hosts[df_hosts['game_location'].str.contains('France', na=False)]

# Nombre d'organisation l'hiver et l'été
jeux_hiver = df_hosts_france[df_hosts_france['game_season'] == 'Winter'].shape[0]
jeux_ete = df_hosts_france[df_hosts_france['game_season'] == 'Summer'].shape[0]

# Ajouter les jeux de 2024 (non présent dans les données)
jeux_ete += 1

print(f"France a organisé {jeux_hiver} Jeux Olympiques d'hiver et {jeux_ete} Jeux Olympiques d'été.")
print(f"Total: {jeux_hiver + jeux_ete} Jeux Olympiques.")
```

```python
# Question 4

paris_1900_results = df_results[df_results['slug_game'] == 'paris-1900']

# Vérifier la présence de femmes dans ces résultats
women_participation = paris_1900_results['event_title'].str.contains('Women', na=False)

#print(women_participation)
if women_participation.any():
    print("Les femmes ont pu participer aux JO de Paris en 1900.")
else:
    print("Les femmes n'ont pas participé aux JO de Paris en 1900.")
```

```python
# Question 5

# Identifier les athlètes ayant remporté des médailles aux JO d'hiver et d'été
medaille_hiver = df_medals[df_medals['slug_game'].str.contains('winter', case=False)]
medaille_ete = df_medals[df_medals['slug_game'].str.contains('summer', case=False)]

# Trouver les athlètes qui ont des médailles dans les deux saisons
atletes_hiver = set(medaille_hiver['athlete_full_name'])
atletes_ete = set(medaille_ete['athlete_full_name'])
athletes_deux_saisons = atletes_hiver.intersection(atletes_ete)

# Vérifier si Christa Ludinger-Rothenburger est parmi eux et vérifier si elle a gagné la même année
if "Christa Ludinger-Rothenburger" in athletes_deux_saisons:
    christa_medaille = df_medals[df_medals['athlete_full_name'] == "Christa Ludinger-Rothenburger"]
    medal_years = christa_medaille['slug_game'].str.extract(r'(\d{4})')[0].astype(int)
    if medal_years.duplicated().any():
        print("Christa Ludinger-Rothenburger a remporté des médailles au cours de la même année.")
    else:
        print("Christa Ludinger-Rothenburger n'a pas remporté des médailles au cours de la même année.")
else:
    print("Christa Ludinger-Rothenburger n'a pas remporté des médailles aux JO d'hiver et d'été.")
```

```python
# Question 6

df_hosts['game_year'] = df_hosts['game_year'].astype(int)

# Filtrer les JO d'été et d'hiver de 1924 à 1992
summer_games_1924_1992 = df_hosts[(df_hosts['game_season'] == 'Summer') & (df_hosts['game_year'] >= 1924) & (df_hosts['game_year'] < 1992)]
winter_games_1924_1992 = df_hosts[(df_hosts['game_season'] == 'Winter') & (df_hosts['game_year'] >= 1924) & (df_hosts['game_year'] < 1992)]

# Filtrer les JO après 1992
summer_games_post_1992 = df_hosts[(df_hosts['game_season'] == 'Summer') & (df_hosts['game_year'] > 1992)]
winter_games_post_1992 = df_hosts[(df_hosts['game_season'] == 'Winter') & (df_hosts['game_year'] > 1992)]

# Vérifier si les JO d'été et d'hiver avaient lieu la même année de 1924 à 1992
same_years_1924_1992 = set(summer_games_1924_1992['game_year']).intersection(set(winter_games_1924_1992['game_year']))

# Vérifier les cycles après 1992
alternating_cycles_post_1992 = all((winter_games_post_1992['game_year'] % 2 == 0) & (summer_games_post_1992['game_year'] % 4 == 0))

# Afficher les résultats
if same_years_1924_1992:
    print("De 1924 à 1992, les JO d’hiver et d’été avaient lieu au cours de la même année.")
else:
    print("De 1924 à 1992, les JO d’hiver et d’été n'avaient pas toujours lieu au cours de la même année.")

if alternating_cycles_post_1992:
    print("Depuis 1994, les JO d’hiver et d’été sont organisés selon des cycles distincts avec une alternance de 2 ans.")
else:
    print("Depuis 1994, les JO d’hiver et d’été ne sont pas toujours organisés selon des cycles distincts avec une alternance de 2 ans.")
```
